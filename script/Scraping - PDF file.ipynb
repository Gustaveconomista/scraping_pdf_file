```python
# Setting and listing files from the respective directory
os.chdir("C:\\Users\\guhhh\\OneDrive\\Área de Trabalho\\Workspace\\scraping_pdf_file\\files\\Guarulhos - SP")
dir = os.listdir()
dir
```

```python
# Scraping
pdf_data = tb.read_pdf(dir[6], pages='all')
pdf_data
```

## Step 3: Clean Dataset

```python
df = pdf_data[0][1:]
df.columns = df.iloc[0]
df = df[0:]
df = df.values.tolist()
lr = pdf_data[1].columns.tolist()
df.append(lr)
for i in range(1, 11):
    dfi = pdf_data[i]
    dfi = dfi.values.tolist()
    lr = pdf_data[i+1].columns.tolist()
    dfi.append(lr)
    for j in range(0, len(dfi)):
        df.append(dfi[j])
dfi = pdf_data[11]
dfi = dfi.values.tolist()
for j in range(0, len(dfi)):
    df.append(dfi[j])
df = pd.DataFrame(df, columns=df[0])[1:]
df.set_index("Ord.", inplace=True, drop=True)
df
```
